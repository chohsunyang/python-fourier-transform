```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, fftfreq, ifft
```

```python
N = 5000
f1, f2, f3, f4, f5 = 90, 180, 290, 350, 700 #frequency in Hz
A1, A2, A3, A4, A5 = 5, 7, 2.6, 4.3, 2.9 #Ampliyude
dt = 2/f1/N
```

```python
t = np.linspace(0, 2/f1, N, endpoint=False)
y = A1*np.sin(2*np.pi*f1*t) + A2*np.sin(2*np.pi*f2*t) + A3*np.sin(2*np.pi*f3*t) + \
    A4*np.sin(2*np.pi*f4*t) + A5*np.sin(2*np.pi*f5*t) #Signal
```

```python
plt.figure(figsize=(15,4), dpi=72)
plt.plot(t, y, linestyle='-', color='blue')
plt.grid()
plt.xlabel('t (s)', fontsize=14)
plt.ylabel('amplitude', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(15,4), dpi=72)
yf = fft(y)
xf = fftfreq(N, dt)
plt.plot(xf[:N//128], 2/N*np.abs(yf[:N//128]), linestyle='-', color='blue')
plt.grid()
plt.xlabel('frequency (Hz)', fontsize=14)
plt.ylabel('intensity', fontsize=14)
```
